```python
import matplotlib.pyplot as plt
import seaborn as sns
# Set a style for better-looking plots
sns.set_style("whitegrid")
```

# Load the CSV

```python
# Load the CSV file into a pandas DataFrame
df = pd.read_csv('/kaggle/input/gdp-per-capita-gov-expenditure-trade/gdp_per_capita.csv')

# Display the first few rows to get a quick look at the data
print("First 5 rows of the DataFrame:")
print(df.head())

# Print some basic information about the DataFrame
print("\nInformation about the DataFrame:")
df.info()
```

# Basic Descriptive Statistics

```python
# Get descriptive statistics for numerical columns
print("\nDescriptive statistics for numerical columns:")
print(df.describe())
```

# Value Counts for Categorical Columns

```python
# Get value counts for categorical columns ('Entity' and 'Code')
print("\nValue counts for 'Entity' column:")
print(df['Entity'].value_counts().head(10)) # Displaying the top 10

print("\nValue counts for 'Code' column:")
print(df['Code'].value_counts().head(10)) # Displaying the top 10
```

# Distribution of GDP per capita (Histogram)

```python
# Visualize the distribution of 'GDP per capita' using a histogram
plt.figure(figsize=(10, 6))
sns.histplot(df['GDP per capita'], kde=True)
plt.title('Distribution of GDP per capita')
plt.xlabel('GDP per capita')
plt.ylabel('Frequency')
plt.show()
```

# Distribution of 'Value of global merchandise exports as a share of GDP' (Histogram)

```python
# Visualize the distribution of 'Value of global merchandise exports as a share of GDP'
plt.figure(figsize=(10, 6))
sns.histplot(df['Value of global merchandise exports as a share of GDP'], kde=True)
plt.title('Distribution of Merchandise Exports as % of GDP')
plt.xlabel('Merchandise Exports (% of GDP)')
plt.ylabel('Frequency')
plt.show()
```
